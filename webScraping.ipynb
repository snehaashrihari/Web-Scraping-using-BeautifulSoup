```python
from bs4 import BeautifulSoup
import requests

url= "https://www.commercialcafe.com/office/us/tx/richardson/"
page= requests.get(url)
print(page)
```

```python
soup= BeautifulSoup(page.content, 'html.parser')
print(soup)
```

```python
lists= soup.find_all('div', class_="results")
for list in lists:
  buildingName= lists.find('h2', class_="building-name").text
  label= lists.find('span', class_="building-label").text
  location= lists.find('span', class_="building-address").text
  info= [buildingName,label,location]
  print(info)
```
